```python
import boto3
import pandas as pd
from io import StringIO, BytesIO
from datetime import datetime, timedelta
```

```python
arg_date='2022-04-08'
```

```python
arg_date_dt=datetime.strptime(arg_date,'%Y-%m-%d').date()-timedelta(days=1)
```

```python
arg_date_dt
```

# Extraction (Read multiple files)

```python
s3=boto3.resource('s3')
bucket=s3.Bucket('deutsche-boerse-xetra-pds')
```

```python
objects=[obj for obj in bucket.objects.all() if datetime.strptime(obj.key.split('/')[0],'%Y-%m-%d').date()>=arg_date_dt ]
```

```python
objects
```

```python
csv_object_init=bucket.Object(key=objects[0].key).get().get('Body').read().decode('utf-8')
data=StringIO(csv_object_init)
df_init=pd.read_csv(data,delimiter=',')
```

```python
df_init
```

```python
df_init.columns
```

```python
df_all=pd.DataFrame(columns=df_init.columns)
for obj in objects:
    csv_object=bucket.Object(key=obj.key).get().get('Body').read().decode('utf-8')
    data=StringIO(csv_object)
    df=pd.read_csv(data,delimiter=',')
    df_all=df_all.append(df,ignore_index=True)
```

```python
df_all
```

```python
csv_object
```

```python
df_all
```

```python
df_init.columns
```

```python
columns=['ISIN', 'Date', 'Time', 'StartPrice', 'MaxPrice', 'MinPrice','EndPrice', 'TradedVolume']
df_all=df_all.loc[:,columns]
```

```python
df_all
```

```python
df_all.isna().count()
```

```python
df_all.dropna(inplace=True)
```

# Transformation

```python
df_all["opening_price"]=df_all.sort_values(by=["Time"]).groupby(['ISIN','Date'])['StartPrice'].transform('first')
```

```python
df_all
```

```python
df_all[df_all['ISIN']=='AT0000A0E9W5']
```

```python
df_all["closing_price"]=df_all.sort_values(by=["Time"]).groupby(['ISIN','Date'])['StartPrice'].transform('last')
```

```python
df_all
```

```python
#Aggregations

df_all=df_all.groupby(['ISIN','Date'],as_index=False).agg(opening_price_eur=('opening_price','min'),closing_price_eur=('closing_price','min'), minimum_price_eur=('MinPrice','min'),maximum_price_eur=('MaxPrice','max'),daily_traded_volumn=('TradedVolume','sum'))
```

```python
df_all
```

```python
#Percentage change in previous day closing price
df_all['Prev_closing_price']=df_all.sort_values(by=['Date']).groupby(['ISIN'])['closing_price_eur'].shift(1)
```

```python
df_all
```

```python
df_all['change_prev_closing_percentage']=(df_all['closing_price_eur']-df_all['Prev_closing_price'])/df_all['Prev_closing_price']*100
```

```python
df_all
```

```python
df_all.drop(columns=['Prev_closing_price'],inplace=True)
```

```python
df_all
```

# Loading 

```python
key='xetra_daily_report'+datetime.today().strftime("%Y%m%d_%H%M%S")+'.parque'
```

```python
out_buffer=BytesIO()
df_all.to_parquet(out_buffer,index=False)
bucket_target=s3.Bucket('xetra-786')
bucket_target.put_object(Body=out_buffer.getvalue(),Key=key)
```
